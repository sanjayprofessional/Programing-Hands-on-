### Step - 2
    to identify the Html tags and elements, exctracting and storing them in variables

#### Exctracted Tags of the Specifics
- Price - (div class) = `_30jeq3 _1_WHN1`

- Rating - (div class)= `_3LWZlK`

- Title - (div class) = `_4rR01T`

- Feature List - (ul class) = `_1xgFaf`

#### -  Finding individual elements to check if they are able to extract

```python
# to extract the elements, we use " .text " and " find " methods.

# price
price = soup.find('div', attrs={'class' : '_30jeq3 _1_WHN1'})

# Titles
title = soup.find('div', attrs={'class' : '_4rR01T'})

# Ratings
rating = soup.find('div', attrs={'class' : '_3LWZlK'})

# Features
feature_list = soup.find('ul', attrs = {'class' : '_1xgFaf'})

# Getting " Print " and " Type " 
print(f"{price.text}\n{title.text}\n{rating.text}\n{feature_list.text}")
print(type(price.text))
print(type(title.text))
print(type(rating.text))
print(type(feature_list.text))

# Desired Outputs 
```

#### - Getting all the URL's and Other elements from tags

```python
# to confirm if we are able to generate all the desired links.

for i in range(1, 61):
    
    # to get all pages of website, we iterate each.
    URL = 'https://www.flipkart.com/search?q=television&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=television%7CTelevisions&requestId=4758dac3-5be5-42c2-acff-86d743d45aa1&as-searchtext=tele&page={}'. format(i)
    print(URL)
    
# as we can see below we are able to get the links and they are totally working.
```

```python
# To collect and extract the elements from the tags and appending to variables.

titles = []
ratings = []
prices = []
features = []

for i in range(1, 61):
    
    # to get all pages of website, we iterate each.
    URL = 'https://www.flipkart.com/search?q=television&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=television%7CTelevisions&requestId=4758dac3-5be5-42c2-acff-86d743d45aa1&as-searchtext=tele&page={}'. format(i)
    
    # to get access to data from the website.
    page = requests.get(URL)
    htmlCode = page.text
    
    # to covert to HTML format using BeautifulSoup.
    soup = BS(htmlCode)
    
    # here we took the whole tag, which contains all the features.
    for container in soup.find_all('div', attrs={'class' : '_2kHMtA'}):

        # from each iteration, we are taking only the title tags and appending with data and Nan.
        product = container.find('div', attrs={'class' : '_4rR01T'})
        if product is None:
            titles.append(np.NaN)
        else:
            titles.append(product.text)

        # from each iteration, we are taking only the title tags and appending with data and Nan.
        price = container.find('div', attrs={'class' : '_30jeq3 _1_WHN1'})
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)

        # from each iteration, we are taking only the title tags and appending with data and Nan.
        rating = container.find('div', attrs={'class' : '_3LWZlK'})
        if rating is None:
            ratings.append(np.NaN)
        else:
            ratings.append(rating.text)

        # from each iteration, we are taking only the title tags and appending with data and Nan.
        feature = container.find('ul', attrs={'class' : '_1xgFaf'})
        if feature is None:
            features.append(np.NaN)
        else:
            features.append(feature.text)
            
# All the required elements from the tags has been succesfully extracted and saved into respectable Variables. 
```

#### spliting the required text from the columns

```python
# using Regular Expressions we extract the specifics.

# Brand
regex = r'^(\w+)\s'
TV['Brand']=TV['Product_name'].apply(lambda x: re.findall(regex,x))

# os
regex1 = r'Operating System:\s(Android|Android Based|Coolita|FireTv|Google TV|HomeOS|Linux|Linux based|Tizen|VIDAA|Web)'
TV['Operating_System']=TV['Features'].apply(lambda x: re.findall(regex1,x))

# Smart_tv
regex2 = r'Smart\s\w+\sTV'
TV['Smart_tv']=TV['Product_name'].apply(lambda x: re.findall(regex2,x))

# Screen_size
regex3 = r'(\d+\s(?:Inch|inch))'
TV['Screen_size']=TV['Product_name'].apply(lambda x: re.findall(regex3,x))

# Screen_type
regex4 = r'\s(LCD|LED|OLED|QLED)'
TV['Screen_type']=TV['Product_name'].apply(lambda x: re.findall(regex4,x))

# Resolution
regex5 = r'\s(Full HD|HD Ready|Ultra HD \(4K\))'
TV['Resolution']=TV['Product_name'].apply(lambda x: re.findall(regex5,x))

# all the required text has been extracted and placed under their respective columns.
```

```python
# here we are coverting the columns type to 'str'.

TV.Brand = TV.Brand.apply(lambda x : ''.join(x))
TV.Operating_System = TV.Operating_System.apply(lambda x : ''.join(x))
TV.Smart_tv = TV.Smart_tv.apply(lambda x : ''.join(x))
TV.Screen_size = TV.Screen_size.apply(lambda x : ''.join(x))
TV.Screen_type = TV.Screen_type.apply(lambda x : ''.join(x))
TV.Resolution = TV.Resolution.apply(lambda x : ''.join(x))

TV

# successfuly converted.
```
